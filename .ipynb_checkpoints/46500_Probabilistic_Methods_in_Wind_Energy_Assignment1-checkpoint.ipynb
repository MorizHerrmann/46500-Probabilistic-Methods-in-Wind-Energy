## [1a] Gumbel Fit - Method of Probability-Weighted Moments (PWM)
The Gumbel parameters $\alpha$ and $\beta$ are calculated from the PWM $b_0$ and $b_1$ for

### Some functions definitions that will be re-used a few times later
The (simplified) extreme wind speed according to a Gumbel distribution

```python
def Vgumbel(alpha, beta, T, T0=1):
    """
    Parameters
    ----------
    alpha : 1st Gumbel parameter.
    beta : 2nd Gumbel parameter.
    T : return period.
    T0 : base period. The default is 1.

    Returns
    -------
    extreme wind speed.

    """
    return beta - alpha * np.log(np.log(1 / (1- T0/T)))
```

Method of Probability-Weighted Moments

```python
def PWM(Umax, T0=1, T50=50, chunksize=0):
    """
    Calculate the extreme wind according to a gumbel distribution which is fit 
    to the data with the probability weighted moments method.

    Parameters
    ----------
    Umax : maxima for each base period (default yearly).
    T0 : base period. The default is 1.
    T50 : return period. The default is 50.

    Returns
    -------
    alpha : gumbel parameter.
    beta : gumbel parameter.
    Vxtr : extreme wind.

    """
    
    if chunksize:
        Umax = np.random.permutation(Umax)
        Umax = np.reshape(Umax, [-1, chunksize])
        
    n = Umax.shape[-1]
    
    # rank the yearly maxima (along last axis)
    UmaxSorted = np.sort(Umax)
    
    # compute moments
    b0 = np.sum(Umax, -1)/n
    b1 = np.sum(np.multiply(np.arange(0, n), UmaxSorted), -1) / n / (n-1)
    
    # compute gumbel parameters
    gamma = 0.577
    alpha = (2*b1 - b0) / np.log(2)
    beta = b0 - gamma * alpha
    
    Vxtr = Vgumbel(alpha, beta, T50, T0=T0)
    
    return alpha, beta, Vxtr, b0, b1
```

Firstly, the $\alpha,\beta$ and $V_{50}$ are calculated from the whole data set.

```python
alpha, beta, V50, b0, b1 = PWM(Umax)

print('PWM: V50= '+str(V50.round(1))+' m/s; alpha='+str(alpha.round(2))+', beta='+str(beta.round(1))+' m/s')

T = np.logspace(np.log10(1.01), np.log10(50))
Vxtr = Vgumbel(alpha, beta, T)

# plot
plt.figure(figsize=(8, 6))
plt.title('Comparison of data with model')
plt.semilogx(TUmax, Umax, 'o', label='ranked yearly maxima')
plt.semilogx(T, Vxtr, label='Gumbel-PWM of whole data')
plt.grid()
tiks = [1, 2, 5, 10, 20, 50]
plt.xticks(tiks, tiks)
plt.xlabel('return period [years]')
plt.ylabel('Extreme wind [m/s]')
plt.legend()
plt.show()

# coefficient of determination
RSS = sum((Umax - Vgumbel(alpha, beta, TUmax))**2)
TSS = sum((Umax - Umax.mean())**2)
Rsq = 1 - RSS/TSS
print(f'Fit of Gumbel distribution to whole data set: RR={Rsq*100:.1f}%')
```

The Gumbel distribution fits the data moderately. The coefficient of determination is ~90%. Especially, in the beginning for return periods smaller than 1.3 years and in the end, when the data points get anyway scarce, there are significant deflection. But for 1.5-10 years the fit seems good. 

Secondly, $\alpha,\beta$ and $V_{50}$ are found using 8 times 2-year non-overlapping chunks and its empirical statistics are computed. The different results are compared in a plot and by the coefficient of determination $R^2$. 

```python
alpha2y, beta2y, V502y, b02y, b12y = PWM(Umax, chunksize=2)
n2y = len(alpha2y)
Vxtr2y = np.empty((len(T), n2y))
for i in range(n2y):
    Vxtr2y[:, i] = Vgumbel(alpha2y[i], beta2y[i], T)

# Because Vgumbel() is linear in alpha and beta the ensemble statistics of the eight Gumbel fits are computed from the the extreme winds and the parameters are bypassed. 
Vxtr2y_mu = np.mean(Vxtr2y, axis=1)
Vxtr2y_std = np.std(Vxtr2y, axis=1)

# plot
plt.figure(figsize=(8, 6))
plt.title('Comparison of data with model')
plt.semilogx(TUmax, Umax, 'o', label='ranked yearly maxima')
plt.semilogx(T, Vxtr, label='Gumbel-PWM of whole data')
#plt.semilogx(Vxtr2y, 'k', alpha=0.3, label='Gumbel-PWM of 2y data')
plt.semilogx(T, Vxtr2y_mu, 'k', label='Mean of 2y Gumbel-PWM')
plt.semilogx(T, Vxtr2y_mu + Vxtr2y_std, 'k--', label='std. uncertainty of 2y Gumbel-PWM')
plt.semilogx(T, Vxtr2y_mu - Vxtr2y_std, 'k--', label='_nolegend_')
plt.grid()
tiks = [1, 2, 5, 10, 20, 50]
plt.xticks(tiks, tiks)
plt.xlabel('Return period [years]')
plt.ylabel('Extreme wind [m/s]')
plt.legend()
plt.show()

print(f'V50 mean deviation of 2y Gumbel-PWM and whole-data Gumbel-PWM: {(Vxtr2y_mu[-1]- Vxtr[-1])/Vxtr[-1]*100:.1f}%')
```

The Gumbel-PWM estimates of the extreme wind fluctuate wildly. The mean has a nice fit to the Gumbel-PWM of the whole data until ~10 years. At the return period the deviation from the original prediction is often high. 

In order to minimize (or better to reveal) the bias of the order of the data set, $U_{max}$ is internally shuffled when calculating the PWM. Thus, the result qualitatively varies when the cell is executed a few times. 

Additionally, the theoretical variance of $V_{50}$ is calculated according to Ott (2011) and will be re-used later. 

```python
def ott2011(alpha, N, T=50, T0=1):
  """
  Calculate the variance of the extreme wind based on a Gumbel distribution according to Ott (2011)

  Parameters
  ----------
  alpha : Gumbel slope parameter
  N : number of base periods
  T : return period. The default is 50.
  T0 : base period. The default is 1.

  Returns
  -------
  var: variance of Gumbel's extreme wind. 

  """
  gamma = 0.577
  qt = np.log(T / T0 - gamma) / np.log(2)
  var = np.power(alpha, 2) * np.pi**2 / 6 / N * (1 + 0.584 * qt + 0.234 * qt**2 / (1 - 0.823 / N))
  return var
```

```python
V502y_var_emp = Vxtr2y_std[-1]**2
V502y_var_ott = ott2011(alpha2y, 2)

plt.figure(figsize=(3, 5))
plt.boxplot(V502y_var_ott/V502y_var_emp)
plt.title('$\sigma^2_{V50, Ott} / \sigma^2_{V50}$')
plt.xticks([])
plt.yscale('log')
plt.grid()
```

The theoretical variance (according to Ott (2011) is often up to one order of magnitude smaller or larger than the empirical one. But over several experiments the ratio is centered around 1. But the model is sensitive to $\alpha$ and there are large and small outliers. 

## Trying different methods to estimate V50
Here, the maximum likelihood method was used to fit a Gumbel and a Weibull distribution. Additionally (and out of curiousity), also the "thumb rule" V_50=5*V_avg was calculated. Results are compared in a plot.

First, the MLE functions are implemented and the negatice log is minimized. 

```python
# 1a : Trying different methods to obtain V50. Here, a maximum likelihood estimate was being used to fit a Gumbel and Weibull distribution

### Using MLE for Gumbel and Weibull
#from scipy import optimize, stats #imported here for portability

# Wsp_sorted = np.sort(Wsp) #Rank the windspeeds # not necessary in this nb

# MLE on the Gumbel distribution 
gumbel_likelihood_max = lambda theta :-np.sum( np.log( abs(sp.stats.gumbel_r.pdf(Umax,loc=theta[0],scale=theta[1]) ) )) # Definition of the negative log-likelihood function; Wsp is the windspeed
gumbel_optimum_max = optimize.minimize(gumbel_likelihood_max,[30,3]) # Minimization to find the optimal distribut=30.1 m/s

# As the Gumbel distribution doesn't look to bad, does Weibull work as well? 
# MLE on the Weibull
weibull_likelihood = lambda theta: -np.sum(np.log(sp.stats.weibull_min.pdf(Umax,c=theta[0],scale=theta[1])))
weibull_optimum = optimize.minimize(weibull_likelihood,[1,1])

print(f' PWM for the Gumbel dist. yields alpha= {alpha}, and beta= {beta}') # adapted to take values from before 
print(f' MLE for the Gumbel dist. yields alpha= {gumbel_optimum_max.x[1]}, and beta= {gumbel_optimum_max.x[0]}')
print(f' MLE for the Weibull dist. yields c = {weibull_optimum.x[0]}, and scale= {weibull_optimum.x[1]}')
```

These were sensitive to the initial conditions. Initial conditions have been changed until results seemed reasonable. 
To visualize the obtained fitted distributions, the MLE distributions are plotted together with the histogram with 10 bins as well as the previously obtained Gumbel distribution fitted using the PWM.

```python
### Plotting of the estimated distributions (pdfs)

#create variable names fitting to the plot
Wsp_sorted = sorted(Wsp) 
UmaxSorted = Umax

fig1 = plt.figure()
axs1 = plt.gca()
hist1 = axs1.hist(UmaxSorted,10, density=True,stacked=True,label="Histogram 10 bins") # 1y max histogram
plt.plot(Wsp_sorted,stats.gumbel_r.pdf(Wsp_sorted,loc = gumbel_optimum_max.x[0],scale = gumbel_optimum_max.x[1]),'r',label="Gumbel MLE") # gumbel mle pdf
plt.plot(Wsp_sorted,stats.gumbel_r.pdf(Wsp_sorted,loc = beta, scale=alpha),color="y",label="Gumbel PWM") # gumbel pwm pdf
plt.plot(Wsp_sorted,stats.weibull_min.pdf(Wsp_sorted,weibull_optimum.x[0],scale=weibull_optimum.x[1]),label="Weibull MLE") # weibull mle pdf
#make the plot look nice
plt.xlabel(r"$U_{max}$")
plt.ylabel("p(u)")
plt.legend()
```

Now the V50 estimates can be calculated and displayed. 

```python
### Finally, calculate V50 estimates
#Calculate V50 values for Gumbell Disribution from the equation in the slides
V50_mle = gumbel_optimum_max.x[0] + gumbel_optimum_max.x[1]*np.log(50) # yields 50 year 
V50PWM = beta + alpha*np.log(50) # yields 50 year 

#get v50 from 2% approximation 
V50_weibull_mle = stats.weibull_min.ppf(.98,weibull_optimum.x[0],scale=weibull_optimum.x[1] )
V50_gumbelPWM_2 = stats.gumbel_r.ppf(.98,loc = beta, scale=alpha) # Sanity check: This should not differ too much from the one calculated by the previous formula ! 

# Out of curiousity, do the thumb rule as well (U50 = 5*U_avg):
V50_thumb = 5* np.mean(Wsp)

#print all results to the output
print(f'V50 for Gumbel with PWM: {V50PWM}')
print(f'V50 for Gumbel with MLE: {V50_mle}')
print(f'V50 for Weibull with MLE: {V50_weibull_mle}')
print(f'V50 for Gumbel with MLE and 2% rule: {V50_gumbelPWM_2}')
print(f'V50 for thumb rule (5* V_avg):  {V50_thumb}')
```

## Testing the Goodness of the fits
#### Using the Kolmogorov-Smirnov-Test (KS-Test)

Here the KS-test is used, to get an impression about how good the previous fits of the distributions are. 

```python
#### Goodness of the fits : Using the KS- test 

#What is the confidence level we need to achieve ? 99% ? 95% ? 
#--> choose 95%. According to http://www.stat.yale.edu/Courses/1997-98/101/confint.htm it is a common choice 

conf_level = .95 # % of confidence; confidence level

# Do the ks test from the scipy.stats module for the three pdf fits 
ks_gumbel_PWM = sp.stats.kstest(UmaxSorted, 'gumbel_r',args=(beta, alpha)) 
ks_gumbel_MLE = sp.stats.kstest(UmaxSorted, 'gumbel_r',args=(gumbel_optimum_max.x[0],gumbel_optimum_max.x[1]),N=16) 
ks_weibull_MLE = sp.stats.kstest(UmaxSorted, 'weibull_min',args=(weibull_optimum.x[0],weibull_optimum.x[1]),N=20)

if False: 
    print(f'KS - Results: \n Gumbel PWM: {ks_gumbel_PWM} \n \
    Gumbel MLE: {ks_gumbel_MLE} \n \
    Weibul Mle: {ks_weibull_MLE}')

#Now compare the p value with the confidence: 
# -> Hypothesis: underlying distribution is the stated one with conf of 95 % 
# Then: Hyp rejected when p < 5
#       Hyp supported when p> 5  # This is according to the examples in scipy.stats.kstest
print("Hypothesis: The Samples follow the regarding distributions with a confidence level of 95%")
if ks_gumbel_PWM[1] >(1-conf_level): 
    print(f"Goodness of Gumbel PWM fit: \t p={np.around(ks_gumbel_PWM[1],4)} is bigger than p_lim={np.around(1-conf_level,4)}, hence the hypothesis is supported. ")
else:
    print(f"Goodness of Gumbel PWM fit: \t p={np.around(ks_gumbel_PWM[1],4)} is smaller than p_lim={np.around(1-conf_level,4)}, hence the hypothesis is rejected. \n \
    The samples don't follow the distributions to the specified confidence.")

    # Same procedure for the Max likelihood estimate
if ks_gumbel_MLE[1] >(1-conf_level): 
    print(f"Goodness of Gumbel MLE fit: \t p={np.around(ks_gumbel_MLE[1],4)} is bigger than p_lim={np.around(1-conf_level,4)}, hence the hypothesis is supported. ")
else:
    print(f"Goodness of Gumbel MLE fit: \t p={np.around(ks_gumbel_MLE[1],4)} is smaller than p_lim={np.around(1-conf_level,4)}, hence the hypothesis is rejected. \n \
    The samples don't follow the distributions to the specified confidence.")

# And again for the mle weibull : 
if ks_weibull_MLE[1] >(1-conf_level): 
    print(f"Goodness of Weibull MLE fit: \t p={np.around(ks_eibull_MLE[1],4)} is bigger than p_lim={np.around(1-conf_level,4)}, hence the hypothesis is supported ")
else:
    print(f"Goodness of Weibull MLE fit: \t p={np.around(ks_weibull_MLE[1],4)} is smaller than p_lim={np.around(1-conf_level,4)}, hence the hypothesis is rejected. \n \
    The samples don't follow the distributions to the specified confidence.")    
```
